1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
def random_list(n, Max):
    return np.random.random(n)*np.random.randint(0,Max,n)

random_list = random_list(10,10)

with open('simple_data.txt', 'w') as f:
    for e in random_list:
        f.write(str(e) + "\n")
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
def random_matrix(Max, n=5):
    return np.random.random((n,n))*np.random.randint(0,Max, (n,n))

matrix = random_matrix(10)

with open('data.txt', 'w') as g:
    for i in range(5):
        for j in range(5):
            g.write(str(matrix[i,j])+' ')
        g.write("\n")  
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
with open('data.txt', 'r') as in_file:
    file_str = in_file.read()
    file_str = file_str.replace(" ", ',')
    print(file_str)
with open('data.csv', 'w') as new_file:
    new_file.write(file_str)
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
with open('credit_card.dat', 'rb') as file: #rb stays for read bit
    word_size = 6
    conv_chr = [] #I will save the converted character in this list
    for line in file:
        for i in range(word_size, len(line)-3, word_size):
            dec = int(line[i-word_size:i], base=2)
            conv_chr += [chr(dec)]

num_chr = 4*4+3 #number of characters within a credit-card number (16 numbers + 3 spaces)

with open('credic_card_numbers.txt', 'w') as new_file:
    for n in np.arange(0, len(conv_chr), num_chr): #for every card number (every raw)
        for i in range(num_chr):                   #for every element of conv_chr for each line
            new_file.write(conv_chr[n+i])
        new_file.write("\n")
```
